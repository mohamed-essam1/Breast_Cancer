#**PCA**

```python
X_meaned = X - np.mean(X , axis = 0)
# print (X_meaned.shape)
```

```python
cov_mat = np.cov(X_meaned , rowvar = False)   # rowvar = False >> work with X_meaned as transposed
```

```python
print (cov_mat.shape)
# print (cov_mat)
```

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

```python
print (eigen_vectors.shape)
# print (eigen_values)
```

```python
#sort the eigenvalues in descending order
sorted_index = np.argsort(eigen_values)[::-1]

sorted_eigenvalue = eigen_values[sorted_index]
#similarly sort the eigenvectors
sorted_eigenvectors = eigen_vectors[:,sorted_index]

print (sorted_eigenvalue)
```

##**Encoding (2)**

```python
n_components = 2 #you can select any number of components.
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
```

```python
print(eigenvector_subset.shape)
```

```python
X_reduced = np.dot(eigenvector_subset.transpose(),X_meaned.transpose()).transpose()
print(X_reduced.shape)
```

#**Fuzzy c means**

```python
class FCM_():
    def __init__(self, n_clusters, max_iter, m, error=0.0000000001):
       self.n_clusters = n_clusters
       self.max_iter = max_iter
       self.m = m
       self.error = error
       self.trained = False


    def fit(self, X):

        n_samples = X.shape[0]
        self.u = np.random.dirichlet(np.ones(self.n_clusters), size=n_samples)
        for _ in range(self.max_iter):
            u_old = self.u.copy()
            self._centers = FCM_._next_centers(X, self.u, self.m)
            self.u = self.soft_predict(X)
            # Stopping rule
            if np.linalg.norm(self.u - u_old) < self.error:
                break
        self.trained = True

    def soft_predict(self, X):

        temp = FCM_._dist(X, self._centers)
        u_ = self.u.copy()

        for i in range(temp.shape[1]):
          u_[:,i] = (1/temp[:,i])**(1/(self.m-1)) / np.sum([((1/temp[:,j])**(1/(self.m-1))) for j in range(temp.shape[1])], axis=0)

        return u_

    def predict(self, X):

        if self.trained:
            # X = np.expand_dims(X, axis=0) if len(X.shape) == 1 else X
            return self.soft_predict(X).argmax(axis=1)

    # def outliers(self, X):
    #   if self.trained:
    #       X = np.expand_dims(X, axis=0) if len(X.shape) == 1 else X

    #   soft_ = self.soft_predict(X)

    #   outliers_ = []
    #   for i in range(len(soft_)):
    #     if np.abs(soft_[i][0] - soft_[i][1]) <= 0.1:
    #       outliers_.append(i)

    #   return outliers_



    def _dist(A, B):
        return np.sqrt(np.einsum("ijk->ij", (A[:, None, :] - B) ** 2))



    def _next_centers(X, u, m):
        """Update cluster centers"""
        um = u ** m
        return (np.dot(X.T, um) / np.sum(um, axis=0)).T

    def centers(self):
        if self.trained:
            return self._centers
```

**FCM Before PCA**

```python
fcm = FCM_(n_clusters=2, max_iter=100, m=1.7)
fcm.fit(X)
```

```python
fcm_centers = fcm.centers()
fcm_softlabels = fcm.soft_predict(X)
fcm_labels_Before = fcm.predict(X)
```

```python
fcm_centers.shape
```

```python
print(fcm_softlabels[:100])
```

**Accuracy of Fuzzy c means**

```python
if accuracy_score(fcm_labels_Before, df.iloc[: , 0])<0.5:
  print(1 - accuracy_score(fcm_labels_Before, df.iloc[: , 0]))
else:
  print(accuracy_score(fcm_labels_Before, df.iloc[: , 0]))
```
